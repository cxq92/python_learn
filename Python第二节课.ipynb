# Group - by  分组计算
plit – apply – combine                         
split: 很具某些条件对数据进行分组                          
apply:对每一个group独立的应用函数  #重点是先分组 再应用函数                    
combine：将结果数据组合到某种数据结构中                   

**重点： Apply               
aggregation: 对每一个小组做一些统计上的计算              
transformation: 对group做一些特别的计算                 
filtration:对group进行某些条件的过滤  **

## merge
joining key columns              
joining key indexes                     
joining key columns on indexes                     


```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A': ['A0', 'A1', 'A2', 'A3'],
                      'B': ['B0', 'B1', 'B2', 'B3']})
 

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                       'key2': ['K0', 'K0', 'K0', 'K0'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=["key1","key2"],how = "inner")#inner指的是两表key相交得到的集合
```

```python
pd.merge(left,right,on=["key1","key2"],how = "left")#以left表的主键为主
```

```python
pd.merge(left,right,on=["key1","key2"],how = "outer") #两张表主键求并的结果
```

```python
left1 = left.rename(columns={"key1":"key1_L","key2":"key2_L"})
```

```python
left1
```

```python
pd.merge(left1,right,left_on = ["key1_L","key2_L"],right_on = ["key1","key2"],how = "inner")
```

#### 对于主键不同的列，不能同之前的方法，直接一个on（，），要分开写left_on，right_on，但是两个表的键要对应，就是说，key1_L与key1相对应。

```python
left2 = left.set_index(["key1","key2"])
```

```python
left2      #  "key1","key2"在这个表的索引上，不是在列上
```

```python
pd.merge(left2,right,right_on=["key1","key2"],left_index=True)
```

#### 左表里的索引与右表里的主键在数量与位置上要一一对应，要求比较严苛。

```python
right2 = right.set_index(["key1","key2"])
```

```python
right2
```

```python
pd.merge(left2,right2,right_index=True,left_index=True)
```

```python
help(pd.merge)
```

## Join
* 组合dataframe 的一种快捷方式，它默认的是在索引上进行Join           
* join有一个参数on，可以指定一个或者多个columns，但是这只对最左边的数据集有效，右边的数据集的key还是索引              

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                     index=pd.Index(['K0', 'K1', 'K2'], name='key'))


index = pd.MultiIndex.from_tuples([('K0', 'Y0'), ('K1', 'Y1'),
                                 ('K2', 'Y2'), ('K2', 'Y3')],
                                  names=['key', 'Y'])


right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3']},
                      index=index)
```

### 如果两张表的主键都在index上，join的功能就很强大！
